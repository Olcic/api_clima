Consumindo API de Clima para entender a temperatura recente e o nome de uma determinada cidade, a partir de coordenadas geográficas.

```python
import requests

# Coordenadas
latitude = -22.91
longitude = -43.17


# Geocodificação Reversa
# Descobre o município pelo Nominatim
url_localizacao = "https://nominatim.openstreetmap.org/reverse"

params_localizacao = {
    "lat": latitude,
    "lon": longitude,
    "format": "json",
    "zoom": 10,
    "addressdetails": 1
}

headers = {
    "User-Agent": "estudo-api/1.0"
}

response_loc = requests.get(
    url_localizacao,
    params=params_localizacao,
    headers=headers
)

if response_loc.status_code != 200:
    print("Erro ao consultar localização")
    municipio = estado = pais = "Desconhecido"
else:
    data_loc = response_loc.json()
    endereco = data_loc.get("address", {})

    municipio = (
        endereco.get("city")
        or endereco.get("town")
        or endereco.get("municipality")
        or "Desconhecido"
    )
    estado = endereco.get("state", "Desconhecido")
    pais = endereco.get("country", "Desconhecido")

# 2) Clima Atual (Open-Meteo)
url_clima = (
    "https://api.open-meteo.com/v1/forecast"
    f"?latitude={latitude}"
    f"&longitude={longitude}"
    "&current_weather=true"
    "&timezone=America/Sao_Paulo"
)

response_clima = requests.get(url_clima)

if response_clima.status_code != 200:
    print("Erro ao consultar clima")
else:
    data_clima = response_clima.json()
    clima = data_clima["current_weather"]

    temperatura = clima["temperature"]
    vento = clima["windspeed"]
    periodo = "Dia" if clima["is_day"] == 1 else "Noite"

    # Resultado Final
    print("Localização:")
    print(f"Município: {municipio}")
    print(f"Estado: {estado}")
    print(f"País: {pais}")
    print()
    print("Clima atual:")
    print(f"Temperatura: {temperatura} °C")
    print(f"Vento: {vento} km/h")
    print(f"Período: {periodo}")
```
